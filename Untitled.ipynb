### Economy Setup (Real, Nominal, Inflation, Exchange, Credit)

```python
#Real yield model
real_yield = economy.SubModel('RealYieldCurves').SubModel('RealYieldCurve') 
real_yield.Type = 'TwoFactorVasicek'

d_SR = {'Alpha': 0.25, 'Sigma': 0.01, 'StartVal': 0.0087} #1 factor Vasicek for ShortRate
d_MRL = {'Alpha': 0.1, 'Sigma': 0.012, 'Mu': 0.0272, 'StartVal': 0.0253} #1 factor Vasicek for MeanReversionLevel 

for i, j in d_SR.items(): real_yield.SubModel('ShortRate').Parameter(i).Value = j
for i, j in d_MRL.items(): real_yield.SubModel('MeanReversionLevel').Parameter(i).Value = j


#Nominal yield model
nominal_yield = economy.SubModel('NominalYieldCurves').SubModel('NominalYieldCurve') 
nominal_yield.Type = 'AnnualLMMPlus'

 #d_SR = {'Alpha': 0.059, 'Sigma': 0.255} #1 factor Black–Karasinski for ShortRate
 #d_MRL = {'Alpha': 0.099, 'Sigma': 0.386} #1 factor Black–Karasinski for MeanReversionLevel 

 #for i, j in d_SR.items(): nominal_yield.SubModel('ShortRate').Parameter(i).Value = j
 #for i, j in d_MRL.items(): nominal_yield.SubModel('MeanReversionLevel').Parameter(i).Value = j

#Inflation = constant
inflation = economy.SubModel('InflationRates').SubModel('Inflation') 
inflation.SubModel('Volatility').Type = 'FixedVolatility'
inflation.SubModel('Volatility').Parameter('Value').Value = 0.0
inflation.Parameter('InitialIndexValue').Value = 1
inflation.Parameter('InitialInflationRate').Value = 0

#Exchange rate model = constant
exchange_rate = economy.SubModel('ExchangeRate')
exchange_rate.SubModel('Volatility').Type = 'FixedVolatility'
exchange_rate.SubModel('Volatility').Parameter('Value').Value = 0.0
exchange_rate.Parameter('StartVal').Value = 0.25

#Credit model (by default)
credit_model = economy.SubModel('CreditModels').SubModel('CreditModel')
credit_model.Type = 'CreditModelG2'
```

### Assets

#### Two major risky assets

```python
assets = sim.FindModelByName('Assets')

#Equity

#correlation is dealt with via the correlation matrix rather than the equity asset factors.
equity_model1 = assets.SubModel('EquityAssets').AddModelWithName('ParentEquityAssetCorrelationModel', 'E_USD')
equity_model2 = assets.SubModel('EquityAssets').AddModelWithName('ParentEquityAssetCorrelationModel', 'E_JPY')
model_list = [equity_model1, equity_model2]

for model in model_list:
    model.SubModel('MeanReturn').Type = 'FixedEquityReturn' #Models the long term mean return dynamics
    model.SubModel('Sigma').Type = 'SVJD' #The volatility structure of the asset: mixture of Heston and the Merton Jump Diffusion
    model.SubModel('Sigma').SubModel('Jump').Type = 'LogNormalJump'
    model.SubModel('Sigma').SubModel('Variance').Type = 'ExtendedOneFactorCIR(J)'  
    model.Parameter('Economy').Value = economy.Name
    model.Parameter('IntraSectorCorrelation').Value = 0.35 #Correlation between assets within this market
```

### Output

```python
#Add desired output parameters
short_rate = nominal_yield.Output('ShortRate') #short rate trajectory for nominal yield
asset1 = equity_model1.Output('TotalReturnIndex') #risky asset 1 trajectory
asset2 = equity_model2.Output('TotalReturnIndex') #risky asset 2 trajectory
bond = FixedI_model.Output('TotalReturnIndex') #bond trajectory
output_list = [short_rate, asset1, asset2, bond]

martingale_test_list = [equity_model1, equity_model2, FixedI_model]
martingale_test_list += risky_models
analysis_list = []
for i in martingale_test_list: analysis_list.append(i.AddAnalysis(i.Analysis('Asset Martingale Test')))
for i in analysis_list: i.Folder = outputPath + '/Martingale test' 
```

```python
name = input()
outputFile = sim.AddOutputFile(name)
outputFile.Folder = outputPath

for i in output_list: outputFile.AddOutput(i)

names = [i.FullName for i in output_list]
names.insert(0, 'Timestep')
names.insert(0, 'Trial')
df_trajectories = pd.DataFrame(columns = names)

df_marting_tests = []
for _ in martingale_test_list: df_marting_tests.append(pd.DataFrame(columns = ['Maturity t', 'EstimatedPrice(t)', 'StandardError(t)', 'ExpectedPrice(t)']))


start = time()
thread_mode = ThreadMode(1)
m =  int(1.5 * 10 ** 1) #number of repetitions
seeds = np.linspace(1, 1000, m)

for i in seeds:
    
    sim.Parameter('RandomNumberSeed').Value = round(i)
    sim.RunLocally(thread_mode, True)

    df1 = pd.read_csv('E:/ScenarioGenerator/Output/'+name+'.csv', index_col = False)
    df_trajectories  = pd.concat([df_trajectories, df1], join = 'inner', ignore_index=True, copy = True)

    for j in range(len(martingale_test_list)):
        path = 'E:/ScenarioGenerator/Output/Martingale Test/' + martingale_test_list[j].Name + '.Asset Martingale Test.csv' 
        df1 = pd.read_csv(path, index_col = False)
        df_marting_tests[j] = pd.concat([df_marting_tests[j], df1], join = 'inner', ignore_index=True, copy = True)
    
df_trajectories.insert(0, 'Seed_num', np.repeat(np.arange(1, m+1, 1), int(sim.Parameter('NumberOfTrials').Value) *  (1 + int(sim.ParameterCollection('Timesteps').get_Value('Annual')))))

column = np.repeat(np.arange(1, m+1, 1), (1 + int(sim.ParameterCollection('Timesteps').get_Value('Annual'))))
for i in df_marting_tests: i.insert(0, 'Seed_num', column)
for i in range(len(martingale_test_list)):
    df_marting_tests[i].name = martingale_test_list[i].Name
end = time()
print(end - start)
```

```python
def plot_g(df):
    seeds = df['Seed_num'].unique()
    
    temp = df[df['Seed_num'] == 1]
    x = temp['Maturity t']
    
    fig, ax = plt.subplots()
    for i in seeds:
        sns.lineplot(x = x, y = df[df['Seed_num'] == i]['EstimatedPrice(t)'].values, label = 'seed '+ str(i))
    plt.xlabel('date')
    plt.ylabel('Value')
    fig.set_figheight(6)
    fig.set_figwidth(9.5)
    plt.title('Martingale test for {} using {} seeds'.format(df.name, len(seeds)))
    plt.show() 
```
