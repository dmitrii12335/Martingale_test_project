```python
#General settings
ESG_dict = {'NumberOfTrials': 10**1 ,  'InitialTrialNumber': 1, 'RandomNumberGenerator': 'MersenneTwister', 
           'UseAntithetics': True, 'RandomNumberSeed': 10**1, 'SeedByTrial': True, 
           'StartTrial': 1, 'UseRiskNeutralValuation': True, 'RebalancingStrategy': 'RebalanceOnMaturity', 
           'MaturityOfNumeraireBond': 1, 'BaseEconomy': economy.Name, 'BaseDate': dt.today().strftime('%d/%m/%Y'), 
           'DependenceStructure': 'GaussianCopula', 'EquityAssetCorrelationMethod': 'CorrelationMatrix',
            'CorrelateEquityJumpShocks': False, 'CreditMigrationFactorMethod': 'EquityZScore', 'StochasticRunType': 'Stochastic'}
for i in ESG_dict.keys(): sim.Parameter(i).Value = ESG_dict[i]

t = sim.ParameterCollection('Timesteps')
t.set_Value('DaysPerYear', 365)
t.set_Value('WeeksPerYear', 52)
for i in ['Daily', 'Weekly', 'Monthly', 'Quarterly', 'SemiAnnual']: t.set_Value(i,0)
t.set_Value('Annual', 120)

print('Correlation matrix is size of {} x {} and is filled with {}'.format(len(list(sim.ParameterMatrix('CorrelationMatrix').get_RowHeaders())), len(list(sim.ParameterMatrix('CorrelationMatrix').get_ColumnHeaders())), set(sim.ParameterMatrix('CorrelationMatrix').GetValues())))

print('The following credit ratings are used {}'.format(list(sim.ParameterSet('CreditClasses').GetColumn('CreditClass'))))
```

### Economy Setup (Real, Nominal, Inflation, Exchange, Credit)

```python
#Real yield model
real_yield = economy.SubModel('RealYieldCurves').SubModel('RealYieldCurve') 
real_yield.Type = 'TwoFactorVasicek'

d_SR = {'Alpha': 0.25, 'Sigma': 0.01, 'StartVal': 0.0087} #1 factor Vasicek for ShortRate
d_MRL = {'Alpha': 0.1, 'Sigma': 0.012, 'Mu': 0.0272, 'StartVal': 0.0253} #1 factor Vasicek for MeanReversionLevel 

for i, j in d_SR.items(): real_yield.SubModel('ShortRate').Parameter(i).Value = j
for i, j in d_MRL.items(): real_yield.SubModel('MeanReversionLevel').Parameter(i).Value = j


#Nominal yield model
nominal_yield = economy.SubModel('NominalYieldCurves').SubModel('NominalYieldCurve') 
nominal_yield.Type = 'AnnualLMMPlus'

 #d_SR = {'Alpha': 0.059, 'Sigma': 0.255} #1 factor Black–Karasinski for ShortRate
 #d_MRL = {'Alpha': 0.099, 'Sigma': 0.386} #1 factor Black–Karasinski for MeanReversionLevel 

 #for i, j in d_SR.items(): nominal_yield.SubModel('ShortRate').Parameter(i).Value = j
 #for i, j in d_MRL.items(): nominal_yield.SubModel('MeanReversionLevel').Parameter(i).Value = j

#Inflation = constant
inflation = economy.SubModel('InflationRates').SubModel('Inflation') 
inflation.SubModel('Volatility').Type = 'FixedVolatility'
inflation.SubModel('Volatility').Parameter('Value').Value = 0.0
inflation.Parameter('InitialIndexValue').Value = 1
inflation.Parameter('InitialInflationRate').Value = 0

#Exchange rate model = constant
exchange_rate = economy.SubModel('ExchangeRate')
exchange_rate.SubModel('Volatility').Type = 'FixedVolatility'
exchange_rate.SubModel('Volatility').Parameter('Value').Value = 0.0
exchange_rate.Parameter('StartVal').Value = 0.25

#Credit model (by default)
credit_model = economy.SubModel('CreditModels').SubModel('CreditModel')
credit_model.Type = 'CreditModelG2'
```

### Assets

```python
assets = sim.FindModelByName('Assets')

#Equity

#correlation is dealt with via the correlation matrix rather than the equity asset factors.
equity_model1 = assets.SubModel('EquityAssets').AddModelWithName('ParentEquityAssetCorrelationModel', '1')
equity_model2 = assets.SubModel('EquityAssets').AddModelWithName('ParentEquityAssetCorrelationModel', '2')
model_list = [equity_model1, equity_model2]

for model in model_list:
    model.SubModel('MeanReturn').Type = 'FixedEquityReturn' #Models the long term mean return dynamics
    model.SubModel('Sigma').Type = 'SVJD' #The volatility structure of the asset: mixture of Heston and the Merton Jump Diffusion
    model.SubModel('Sigma').SubModel('Jump').Type = 'LogNormalJump'
    model.SubModel('Sigma').SubModel('Variance').Type = 'ExtendedOneFactorCIR(J)'  
    model.Parameter('Economy').Value = economy.Name
    model.Parameter('IntraSectorCorrelation').Value = 0.35 #Correlation between assets within this market

#Set correlation between two equities = 0.35 in the correlation matrix
sim.ParameterMatrix('CorrelationMatrix').SetValueByName(equity_model1.FullName+'.dZ', equity_model1.FullName+'.dZ', 0.35 )

#FixedIncome

# portfolio asset composed of bonds that are subject to default
FixedI_model = assets.SubModel('FixedIncome').SubModel('GenericBondPortfolios').AddModel('RiskyGenericBondPortfolio')

d_FI = {'Strategy': 'Fixed coupon and maturity', 'Economy': economy.Name, 'NominalYieldCurve': nominal_yield.FullName,
       'InflationRates': inflation.FullName, 'CreditModel': credit_model.FullName, 'EquityAsset': equity_model1.FullName, 
       'CreditClass': 'BB', 'BondType': 'Nominal', 'NumberOfBonds': 1, 'InfiniteNumberOfBonds': False, 'Coupon': 0.05, 
       'Frequency': 1, 'Seniority': 3, 'Term': 30, 'TermDeclineRate': 0.0, 'Duration': 10, 'Convexity': 100}
for i, j in d_FI.items(): FixedI_model.Parameter(i).Value = j
```
